```python
import pandas as pd

df = pd.read_csv("C:/Users/priya/OneDrive/Desktop/adult-income-project/data/adult/cleaned_adult.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
df["age"].hist(bins=30)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
df["hours-per-week"].hist(bins=30)

plt.title("Hours per Week Distribution")
plt.xlabel("Hours per Week")
plt.ylabel("Count")
plt.show()
```

```python
df["income"].value_counts().plot(kind="bar")

plt.title("Income Distribution (0 = <=50K, 1 = >50K)")
plt.xlabel("Income Class")
plt.ylabel("Count")
plt.show()
```

```python
df.groupby("education")["income"].mean().sort_values().plot(kind="bar")

plt.title("Probability of >50K Income by Education Level")
plt.ylabel("Chance of earning >50K")
plt.show()
```

```python
df.groupby("sex")["income"].mean().plot(kind="bar")

plt.title("Probability of >50K Income by Sex")
plt.ylabel("Chance of earning >50K")
plt.show()
```

```python
df.groupby("marital-status")["income"].mean().sort_values().plot(kind="bar")

plt.title("Probability of >50K Income by Marital Status")
plt.ylabel("Chance of earning >50K")
plt.show()
```

# 02  Exploratory Data Analysis (EDA)

The purpose of this notebook is to understand the structure, distribution, and relationships within the cleaned Adult Census dataset. EDA helps identify patterns, correlations, and potential issues prior to model development.

---

## 1. Load Cleaned Dataset

The cleaned dataset (`cleaned_adult.csv`) generated in the Data Preparation notebook is loaded here for visualization and statistical exploration.

---

## 2. Univariate Analysis (Understanding Single Columns)

###  Age Distribution
A histogram was plotted to observe how ages are distributed across the population.  
Key observations:
- Most individuals fall between 25 and 50 years old.
- Few participants are under 20 or over 70.

---

###  Hours-per-Week Distribution
This histogram reveals:
- A strong concentration around **40 hours per week**, which aligns with full-time employment.
- Some individuals work significantly more or fewer hours.

---

###  Income Class Distribution
The target variable shows:
- Income 50K (label: 0) is far more common than >50K (label: 1).
- This indicates **class imbalance**, which affects model training.

---

## 3. Bivariate Analysis (Comparing Features to Income)

###  Income by Education Level
A bar chart of mean income per education group showed:
- Higher education correlates strongly with higher income.
- Doctorate and Masters have the highest earnings probability.

---

###  Income by Sex
Grouping by `sex` highlighted:
- Males have a higher probability of earning >50K.
- Represents gender-based earnings disparity in the dataset.

---

###  Income by Marital Status
Insights:
- Married-civ-spouse has the highest earning probability.
- Never-married and separated groups show much lower income probability.

---

## 4. Key Insights Summary

- The dataset is **imbalanced**, with fewer high-income individuals.
- Age, education level, and hours worked per week are significant factors.
- Categorical variables such as sex and marital-status show noticeable income differences.
- These insights will help guide feature engineering and modeling decisions.

---

##  EDA Output
A deeper understanding of the datasets patterns and relationships, which informs the modeling strategy in the next notebook.


### EDA Summary

- Younger individuals tend to earn less.
- Working more hours per week generally increases income.
- Dataset is imbalanced: fewer people earn above 50K.
- Higher education levels show higher probability of earning >50K.
- Males show higher income probability on average.
- Married individuals (especially "Married-civ-spouse") have higher income probability.
